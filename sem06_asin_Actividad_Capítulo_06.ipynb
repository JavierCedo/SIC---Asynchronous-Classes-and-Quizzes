```python
# Espacio disponible para desarrollar el código y agregar los comentarios necesarios.

'''
Un algoritmo Greedy, o avaro, es una estrategia que toma la mejor opción
disponible en cada paso, sin considerar las consecuencias a largo plazo.
Se enfoca en obtener un resultado óptimo en el corto plazo, con la esperanza
de que esta elección conduzca a una solución global óptima. Aunque esto no
siempre es garantía de obtener la solución óptima global, es eficiente y
fácil de implementar.

Ventajas:

Simplicidad: Son fáciles de entender y programar.

Eficiencia: Suelen tener tiempos de ejecución rápidos, ya que no evalúan
todas las posibles soluciones.

Desventajas:

Soluciones subóptimas: No siempre garantizan la solución óptima global, ya
que se enfocan en la mejor opción local.

Dependencia del problema: Su efectividad depende de la estructura específica
del problema en cuestión.
'''

def coin_change(coins, amount):
    coins.sort(reverse=True)
    change = []
    for coin in coins:
        while amount >= coin:
            amount -= coin
            change.append(coin)
    return change

coins  = [11, 5, 1]
amount = int(input("Ingrese la cantidad: "))
changes= coin_change(coins, amount)
print(changes, len(changes))

coins  = [11, 5, 1, 15, 3, 4]
amount = int(input("Ingrese la cantidad: "))
changes= coin_change(coins, amount)
print(changes, len(changes))

coins  = [1, 3, 5, 7, 9, 11, 13, 15]
amount = int(input("Ingrese la cantidad: "))
changes= coin_change(coins, amount)
print(changes, len(changes))

'''
Para estas entradas el algoritmo busca primero igualar el monto con menedas
desde las mas grandes hasta las mas pequeñas, con esto verifica sin el 11 es
mayor al monto, como no lo es resta ese numero al munto quedando 4, tambien
lo hace con el 5 pero como esta moneda es mayor salta a la siguiente tipo de
moneda siendo la de 1 volviendo a restar poco a poco hasra llegar a cero,
almacenanodo lo valores y cuantas monedas se han usado.
'''
```

# Actividad: Replique el código de la búsqueda de la trayectoria triangular mínima.


## Instrucciones

```python
# Espacio para desarrollar el código y agregar los comentarios necesarios.

def find_minimum(row, col, triangle):
  # primero se verifica si la fila donde se comienza es igual al numero de
  # filas del triangulo.
  if row == len(triangle):
    return 0    # Si solo hay una fila no existe suma y devuelve 0
  else:
    # Se aplica el codigo de forma recursiva asi verificando cada
    # fila pero en un valor de columna distinto para buscar el de menor valor.
    minimum = min(find_minimum(row + 1, col, triangle),
                  find_minimum(row + 1, col + 1, triangle))
    return triangle[row][col] + minimum     # Finalmente se suman los
                                            # valores nimimos obtenidos.

triangle = [
    [2],
    [3, 4],
    [6, 5, 7],
    [4, 1, 8, 3]
    ]
minimum = find_minimum(0, 0, triangle)
print('This minimum cost is ', minimum)

triangle = [
    [2],
    [3, 4],
    [6, 5, 7],
    [4, 1, 8, 3],
    [4, 1, 8, 3, 4],
    [4, 1, 8, 3, 4, 5],
    [4, 1, 8, 3, 4, 5, 6]
    ]
minimum = find_minimum(0, 0, triangle)
print('This minimum cost is ', minimum)
```

# Actividad: Búsqueda binaria con iteración.


## Instrucciones


1. Replique el código mostrado en la unidad 31.

3. Explique por qué es necesario usar usar una función recursiva.

4. Haga diferentes pruebas cambiando la entrada.


```python
# Espacio para desarrollar el código y agregar los comentarios necesarios.

def bin_search(nums, x):
  low, high = 0, len(nums) - 1
  while low <= high:
    mid = (low + high) // 2
    if nums[mid] == x:
      return mid
    elif nums[mid] > x:
      high = mid - 1
    else:
      low = mid + 1
  return -1

'''
El codigo se basa en escencia en una busqueda binaria que ya es un concepto ya
explorado, lo que hace es dividir la lista en dos lista y buscar, si esta en
una lista donde no esta el elemento divide la otra y empieza a buscar, asi
hasta encontrar el elemento.

Se necesita la recursividad para por hacer suboperaciones en elemneto analisados
ya en una operacion atras, y asi constante mente generando una geranquia de
operaciones que depneden de la anterior.
'''

nums = [2, 3, 10]
x = 10
result = bin_search(nums, x)
print(result)

nums = [ 2, 3, 4, 10, 40, 50, 60, 70, 80, 90, 100, 110, 120, 130,
        140, 150, 160, 170]
x = 90
result = bin_search(nums, x)
print(result)

```
