# Actividad: Ordenamiento de Burbuja


## Instrucciones

1. Explique brevemente y en forma clara cómo funciona el algoritmo de ordenamiento de burbuja.

2. Desarrollo un función donde se use el ordenamiento de burbuja aplicada a una lista de valores enteros.



```python
# Espacio para desarrollar el código requerido, recuerde agregar algunos
# comentarios breves sobre su funcionamiento.



# El algoritmo de ordenamiento de burbuja funciona comparando pares de
# elementos adyacentes en una lista y los intercambia si están en el orden
# incorrecto. Este proceso se repite varias veces hasta que no se realizan más
# intercambios, lo que indica que la lista está ordenada.


# Ahora se tiene una implementación de la función que usa el ordenamiento de
# burbuja para ordenar una lista de valores enteros:

def bubbleSort(S):
    n = len(S)
    for i in range(n):
        for j in range(n - 1 - i):
            if S[j] > S[j + 1]:
                S[j], S[j + 1] = S[j + 1], S[j]
    return S


# Ejemplo de uso

lista = [50, 30, 40, 10, 20]
print("Lista original:", lista)
lista_ordenada = bubbleSort(lista)
print("Lista ordenada:", lista_ordenada)

```

```python
# Espacio para desarrollar la actividad, si tienes alguna duda o comentario adicional
# lo puedes dejar planteado acá.



def swap_elements(S, elm1, elm2):
    # Encuentra los índices de los elementos elm1 y elm2
    index1 = S.index(elm1)
    index2 = S.index(elm2)
    # Intercambia los elementos en las posiciones index1 e index2
    S[index1], S[index2] = S[index2], S[index1]

# Ejemplo de uso
lista = [1, 2, 3, 4, 5]
print("Lista original:", lista)
swap_elements(lista, 1, 3)
print("Lista después del intercambio:", lista)


# La función swap_elements intercambia dos elementos en una lista según los
# índices proporcionados. Primero, toma la lista y los índices de los elementos
# a intercambiar. Luego, realiza el intercambio y actualiza la lista


```

```python
# Celda para desarrollar


# Quick Sort es un algoritmo eficiente de ordenamiento que utiliza la
# técnica de "divide y vencerás". Funciona eligiendo un "pivote" de la
# lista y reordenando los elementos para que todos los menores que el
# pivote queden a su izquierda y los mayores a su derecha. Luego, se aplica
# recursivamente el mismo procedimiento a las sublistas de elementos
# menores y mayores que el pivote, hasta que toda la lista esté ordenada.



def partition1(S, low, high):
    pivot = S[low]
    left , right = low + 1, high
    while left < right:
      print(S)
      while left <= right and S[left] <= pivot:
          left += 1
      while left <= right and S[right] >= pivot:
          right -= 1
      if left < right:
        S[left], S[right] = S[right], S[left]
    pivotpoint = right
    S[low], S[pivotpoint] = S[pivotpoint], S[low]
    return pivotpoint

def quickSort1(S, low, high):
    if low < high:
      print(S)
      pivotpoint = partition1(S, low, high)
      quickSort1(S, low, pivotpoint - 1)
      quickSort1(S, pivotpoint + 1, high)


# Ejemplo de uso
S = [6, 2, 11, 7, 5, 4, 8, 16, 10, 3]
quickSort1(S, 0, len(S) - 1)
print("Lista ordenada:", S)


```
