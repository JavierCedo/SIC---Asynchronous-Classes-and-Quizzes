## Q 30-01

Este es el problema del intercambio de monedas.

Tienes una moneda de 400 wones.

Escriba el resultado de cómo el algoritmo `coin_change()` determina el cambio de la moneda de 800 wones.
```
coins  = [500, 400, 100, 50, 10]
amount = int(input("Ingrese la cantidad: "))
changes= coin_change(coins, amount)
print(changes, len(changes))

Ingrese la cantidad: 800
```

````python
def coin_change(coins, amount):
    coins.sort(reverse=True)
    change = []
    for coin in coins:
        while amount >= coin:
            amount -= coin
            change.append(coin)
    return change

coins  = [500, 400, 100, 50, 10]
amount = int(input("Ingrese la cantidad: "))
changes= coin_change(coins, amount)
print(changes, len(changes))
````

Finalmente se muestra la lista de la monedas que corresponden al cambio y ademnas se muestra el numero de cuantas son.

## Q 31-01

Tienes 8 monedas idénticas numeradas del 1 al 8. De estas, solo una moneda es más pesada que la otra.

Suponiendo que puede pesar monedas con una balanza equilibrada, diseñe un algoritmo que encuentre la moneda más pesada.

En este momento, ¿cuántas veces se debe usar al menos la balanza?

````python
coins = {1:1, 2:1, 3:2, 4:1, 5:1, 6:1, 7:1, 8:1}

def balanza(coins):
  for coin in coins:
    peso_mayor = coins[1]
    if coins[coin] > peso_mayor:
      print('Este peso de la moneda', coin, 'es el mayor.  !!!!')
      peso_mayor = coins[coin]
    else:
      print('Este peso de la moneda', coin, 'no en el mayor')

balanza(coins)
````

El codigo lo que hace es apartir de un diccionario con las monedas y sus pesos comparar el peso de la moneda actual con el peso minimo (en primera instancia) de una moneda guardado en la variable peso_mayor, en la comparacion de si el valor del peso de la moneda es igual al guardado en peso_mayor solo imprime que no es la moneda con el peso mas grande, en cambio si en la comparacion de si el valor del peso de la moneda es mayor al guardado en peso_mayor se imprime que es la moneda con el peso mas grande y se actualiza el valor de la variable, asi pudiendose determinar cual es la moneda mas pesada.

La balanza se usa 8 veces ya que se parte de un peso minimo para poder compara todas las monedas.
No logro hacer que se compare directamente sin partir de un valor minimo.

````python
import time

N = int(input("Ingrese un numero: "))

def fib1(n):
  if n==0 or n==1:
    return n
  else:
    return fib1(n-1) + fib1(n-2)

F1 = {0:0, 1:1}

def fib2(n):
  if n in F1:
    return F1[n]
  else:
    F1[n] = fib2(n-1) + fib2(n-2)
    return F1[n]

F2 = [0, 1] + [-1] * (N -1)

def fib3(F, n):
  if n <= 1:
    return n
  else:
    if F[n] < 0:
      F[n] = fib3(F, n-1) + fib3(F, n-2)
    return F[n]


def fib4(n):
    if n <= 1:
        return n
    else:
        a,b = 0,1
        for i in range(2, n + 1):
            a,b = b,a+b
        return b


start1 = time.time()
print('fib1')
print(fib1(N))
end1 = time.time()
print("Elapsed Time:", end1 - start1, "seconds")

start2 = time.time()
print('fib2')
print(fib2(N))
end2 = time.time()
print("Elapsed Time:", end2 - start2, "seconds")

start3 = time.time()
print('fib3')
print(fib3(F2, N))
end3 = time.time()
print("Elapsed Time:", end3 - start3, "seconds")

start4 = time.time()
print('fib4')
print(fib4(N))
end4 = time.time()
print("Elapsed Time:", end4 - start4, "seconds")
````

Analizando para diferentes valores de N, el algoritmo mas lento es fib1. Luego fib2, fib3y fib4 se comportan muy similares, aveces destaca uno mas que otro pero de media general son similares los tiempos.

## Q 32-02

Analice el resultado de la ejecución del siguiente código y compárelo con el rendimiento de las funciones `bin1()` y `bin2()`.
```

```

````python
import time

N = int(input("Ingrese un numero: "))

def bin1(n,k):
    if k==0 or k==n:
        return 1
    else:
        return bin1(n-1,k-1) + bin1(n-1,k)

def bin2(n,k):
    B = [[0]*(i+1) for i in range(n+1)]
    for i in range(n+1):
        for j in range(i+1):
            if j==0 or j==i:
                B[i][j] = 1
            else:
                B[i][j] = B[i-1][j-1] + B[i-1][j]
    return B[n][k]

def bin3(n,k):
    B = [0] * (n + 1)
    for i in range(n + 1):
        for j in range(n, -1, -1):
            if j==0 or j==i:
                B[j] = 1
            else:
                B[j] = B[j], B[j-1]
    return B[k]

start = time.time()

for i in range(N):
    for j in range(i+1):
        print(bin1(i,j),end=" ")
    print()

end = time.time()
print("Elapsed Time:", end - start, "seconds")

start = time.time()

for i in range(N):
    for j in range(i+1):
        print(bin2(i,j),end=" ")
    print()

end = time.time()
print("Elapsed Time:", end - start, "seconds")

start = time.time()

for i in range(N):
    for j in range(i+1):
        print(bin3(i,j),end=" ")
    print()

end = time.time()
print("Elapsed Time:", end - start, "seconds")
````

Analizando para diferentes valores de N, el algoritmo mas rapido es bin2. Luego bin1 y bin3 se comportan muy similares, aveces destaca uno mas que otro pero de media general son similares los tiempos.

## Q 33-01

Si reemplaza cada letra con un número en la siguiente letra, el número correspondiente a las cinco palabras son todos cuadrados.

Para cada palabra, la suma de cada dígito también es un número cuadrado.

Encuentra el número representado por cada letra.

**Código de referencia y Respuesta:**

````python
from math import sqrt, floor
import time

def is_square_digitsum(n):
    # Con esta funcion se ferifica si el numero es un cuadrado perfecto de
    # la suma de digitos.
    s = 0
    while n > 0:
        s += n % 10
        n = n // 10
    if sqrt(s) == int(sqrt(s)):
        return True
    return False

def find_all_squares():
    sqrs = [[] for i in range(5)]
    # lista de listas para almacenar los cuadrados.
    for i in range(1, floor(sqrt(10 ** 5)) + 1):
        n = i * i
        if not is_square_digitsum(n):         # verifica si no es un cuadrado
                                              # perfecto de la suma de digitos.
            continue
        # Se organizan los numero generar una lista de cuadrados perfectos que
        # cumplen ciertas condiciones específicas.
        s = str(n)
        if len(s) == 3 and s[1] != s[2]:
            continue
        if len(s) == 5 and s[2] != s[3]:
            continue
        if len(s) in [4, 5] and len(set(s)) != 4:
            continue
        # Si se cumplen estas condiciones se agrega al numero a una lista donde
        # las longitudes de los numeros son iguales.
        sqrs[len(s) - 1].append(n)
    return sqrs

def promising(s, n, dic):
    # Esta funcion promete que cada letra tenga un digito asignado de
    # la cantidad.
    for i in range(len(s)):
        if s[i] in dic and dic[s[i]] != int(str(n)[i]):
            return False
        # verifica si el digito ya tiene valores asignados y si la letra no
        # esta asociada.
        if int(str(n)[i]) in dic.values() and s[i] not in dic:
            return False
    return True

def solve(words, dic, squares, solved):
    if len(words) == 0:
        return True
    else:
        # cada palabra se intenta ligar con una cantidad de los cuadrados.
        s = words[0]
        candidates = squares[len(s) - 1]
        for n in candidates:
            # por cada cantidad en una lista de su logitud se verifica
            # la promesa.
            if promising(s, n, dic):
                # se crea un nuevo diccionario.
                newdic = dic.copy()
                # se le asigna a cada letra un digito.
                for i in range(len(s)):
                    newdic[s[i]] = int(str(n)[i])
                # Se sigue una recursion asi para cada palabra.
                if solve(words[1:], newdic, squares, solved):
                    solved.update(newdic)
                    return True
        return False

def main():
    # se hace una lista de los cuadrados.
    squares = find_all_squares()
    # lista de palabras de la oracion.
    words = ['A', 'TO', 'ALL', 'XMAS', 'MERRY']
    dic = {}
    solved = {}
    # funcion para resolver el problema.
    solve(words, dic, squares, solved)
    for word in words:
        # se empareja cada palabra con los dijitos de cada letra de esta.
        print(word, end=": ")
        for c in word:
            print(solved[c], end="")
        print()

start = time.time()
main()
end = time.time()
print("Elapsed Time:", end - start, "seconds")
````
