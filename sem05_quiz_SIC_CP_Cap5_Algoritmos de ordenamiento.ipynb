¿Cuántas comparaciones se ejecutaron en el siguiente proceso de ordenamiento de burbuja?
```
def bubbleSort(S):
    n = len(S)
    for i in range(n):
        print(S)
        for j in range(n-1):
            if S[j] > S[j+1]:

S = [50, 30, 40, 10, 20]
bubbleSort(S)
```

```
def insertionSort2(S):
    n = len(S)
```

## Q 28-01

¿Cuántas veces se ejecuta la función `merge2()` en el siguiente proceso de ordenamiento por fusión?
```
S = [6,2,11,7,5,4,8,16,10,3]
mergeSort2(S, 0, len(S)-1)
print(S)
```

````python
def merge2(S, low, mid, high):
    R = []
    print('Ejecucion de merge2')
    i, j = low, mid + 1
    while i <= mid and j <= high:
        if S[i] <= S[j]:
            R.append(S[i])
            i += 1
        else:
            R.append(S[j])
            j += 1
    while i <= mid:
        R.append(S[i])
        i += 1
    while j <= high:
        R.append(S[j])
        j += 1
    for k in range(len(R)):
        S[low + k] = R[k]

def mergeSort2(S, low, high):
    if low < high:
        print(S)
        mid = (low + high) // 2
        mergeSort2(S, low, mid)
        mergeSort2(S, mid + 1, high)
        merge2(S, low, mid, high)

S = [6, 2, 11, 7, 5, 4, 8, 16, 10, 3]
mergeSort2(S, 0, len(S) - 1)
print("Lista ordenada:", S)
````

**Respuesta:**

Poniendo un "contador" en el codigo resulta:

Para la lista con 10 elementos, la funcion merge2() se ejecura 9 veces.

## Q 29-01

Dada la lista a continuación, escriba la salida después de ejecutar la función `partition1()`
```
S = [15,10,12,20,25,13,22]
partition1(S, 0, len(S)-1)
print(S)
```
#### Output
```
[15,10,12,20,25,13,22] 0 6 pivot = 15
...
```


````python
def partition1(S, low, high):
    pivot = S[low]
    left, right = low + 1, high
    print(S, left, right, 'pivot =', pivot)

    while True:
        while left <= right and S[left] <= pivot:
            left += 1
        while left <= right and S[right] >= pivot:
            right -= 1
        if left < right:
            S[left], S[right] = S[right], S[left]
        else:
            break

    pivotpoint = right
    S[low], S[pivotpoint] = S[pivotpoint], S[low]
    print(S, left, right, 'pivot =', pivot)
    return pivotpoint

def quickSort1(S, low, high):
    if low < high:
        pivotpoint = partition1(S, low, high)
        quickSort1(S, low, pivotpoint - 1)
        quickSort1(S, pivotpoint + 1, high)

# Ejemplo de uso
S = [15, 10, 12, 20, 25, 13, 22]
quickSort1(S, 0, len(S) - 1)
print("Lista ordenada:", S)
````
